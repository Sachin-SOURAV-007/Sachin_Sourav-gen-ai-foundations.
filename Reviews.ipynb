```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("Reviews.csv", nrows=50000)

print("Rows Loaded:", len(df))
df.head()
```

```python
light_reviews = df[
    df["Text"].str.contains(r"\blight\b", case=False, na=False)
]["Text"]

print("Number of reviews containing 'light':", len(light_reviews))

for i, review in enumerate(light_reviews.head(20)):
    print(f"\nReview {i+1}:")
    print(review)
```

```python
sentences = [
    "If you are looking for less caramel flavor, you'll need to use the light version of this product (equally good in all applications).",

    "Kettle Brand potato chips are unmistakable--a light gold color, rich flavor and amazing crunch.",

    "These chips are truly amazing. They're light, crisp, great tasting, nice texture, AND they're all natural."
]
```

```python
vectorizer = CountVectorizer()

dtm = vectorizer.fit_transform(sentences)

dense_matrix = dtm.toarray()

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nDense Matrix:")
print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("Vector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)

print("Similarity (Sentence 1 vs Sentence 2):", similarity_1_2)
print("Similarity (Sentence 1 vs Sentence 3):", similarity_1_3)
```

```python
print("\nShared words between Sentence 1 and Sentence 2:")

words = vectorizer.get_feature_names_out()

for i in range(len(words)):
    if vec_1[i] > 0 and vec_2[i] > 0:
        print(words[i])
```
